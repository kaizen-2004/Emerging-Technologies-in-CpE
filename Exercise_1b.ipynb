```python
with open("/home/steve/Python/Exercise_1/hamlet.txt", "r", encoding="utf-8") as f:
    text = f.read()
```

```python
# 1. Tokenize the data using Penn Tree Bank Tokenization
```

```python
tokens = TreebankWordTokenizer().tokenize(text)
```

```python
print(tokens)
```

```python
# 2. Apply case folding on the tokens.
```

```python
case_folded = [token.lower() for token in tokens]
```

```python
print(case_folded)
```

```python
# 3. Lemmatize using WordNet Lemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in case_folded]
```

```python
print(lemmatized_tokens)
```

```python
# 4. Calculate metrics
```

```python
# a. Word types (Number of unique tokens)
```

```python
word_types = len(set(lemmatized_tokens))
```

```python
print(f"Number of unique tokens: {word_types}")
```

```python
# b. Word tokens (Number of tokens)
```

```python
word_tokens = len(lemmatized_tokens)
```

```python
print(f"Total number of tokens: {word_tokens}")
```
